```python

#Code for Percentage Identity
#Percentage identity tells us how similar the two sequences are.
from Bio.Blast import NCBIXML

with open("my_blast_result.xml") as result_handle:
    blast_records = NCBIXML.parse(result_handle)

    for record in blast_records:
        for alignment in record.alignments:
            for hsp in alignment.hsps:
                # Calculate Percentage
                percent_identity = (hsp.identities / hsp.align_length) * 100

                print(f"Match: {alignment.title}")
                print(f"Percentage Identity: {percent_identity:.2f}%") # .2f rounds to 2 decimals
```

Code for Number of Differences
"Differences" usually mean anywhere the sequences don't match. This includes mismatches (wrong base) and gaps (missing base).
formula -  Alignment Length - Identities = Total Differences

```python
from Bio.Blast import NCBIXML

with open("my_blast_result.xml") as result_handle:
    blast_records = NCBIXML.parse(result_handle)

    for record in blast_records:
        for alignment in record.alignments:
            for hsp in alignment.hsps:
                # Calculate Differences
                # align_length includes matches, mismatches, and gaps
                total_differences = hsp.align_length - hsp.identities

                print(f"Match: {alignment.title}")
                print(f"Total Differences: {total_differences}")
                print(f"Gaps in alignment: {hsp.gaps}") # Optional: specifically shows gaps
```

hsp.identities: Eta holo koyta character ekdom mile gese (Perfect matches).

hsp.align_length: Eta holo total length (Matches + Mismatches + Gaps).

Calculation Logic: Percentage ber korar somoy amra total length diye bhag kori. Ar differences ber korar somoy total length theke mil (identities) gulo bad diye dei.

Formatting (:.2f): Python-e floating point number (doshomik) onek lomba hote pare. :.2f use korle eta sudhu doshomik-er por 2 ghor porjonto dekhabe, jeta dekhte clean lage.

```python
from Bio import SeqIO
from Bio import pairwise2

sequence = []

for records in SeqIO.parse("FASTA1-1.fasta", "fasta"): #["sequence no 1 ", "sequence no 2 "....]
    sequence.append(str(records.seq))

seq_1 = sequence[0]
seq_2 = sequence[1]


def global_align(s_1, s_2):
    for a in pairwise2.align.globalxx(s_1, s_2, one_alignment_only = True):
        alignment = pairwise2.format_alignment(*a)
    return alignment

print(global_align(seq_1, seq_2))

```

```python
names_list = ['Jim', 'Jeff', 'Geoff', 'Grace', 'George', 'Julie', 'Justine', 'Jemima'] #defining  lists of names
age_list = ['22', '34', '46', '58', '70', '19', '35', '64']  #defining list of age

""" ekta function thakbe jeitar kaj hobe 2 ta list theke item gula ke niye
serialized ekta dictionary banabe jeikhnae amar 2 ta list er item gula pair and value
hishebe niye amra dictionary return korbe

 """

def combine_lists(list1, list2):  #defining the function
    if len(list1) != len(list2):  #we are checking if the length of the list is equal. otherwise it will throw us error
        raise ValueError("Lists must have the same size") #we are handling the exception here and returning with a value error
    else:    #we are going to this part if the IF statement condition is false
        comb_dict = {}   #declare an empty dictionary
        for x in range(0, len(list1), 1):  #telling the program to run loop
            comb_dict[list1[x]] = (list2[x])  #setting the key and the values of the dictionary
    return comb_dict #returning the updated dictionary

combine_lists(names_list, age_list)
```

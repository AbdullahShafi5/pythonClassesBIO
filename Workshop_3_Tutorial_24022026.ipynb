```python
## Q4 Annotate the code

def reverse(s):
    '''Give a string or other data class (s), returns a string (or other class) reversed.'''
    return (s[::-1])
def complement(s):
    '''Give a sting (must be nucleotides or AA), returns a string with the complement residues.'''
    #create the complement dictionary
    basecomplement = {'A': 'T', 'C': 'G', 'G': 'C', 'T': 'A', 'a': 't', 'c': 'g',
'g': 'c', 't': 'a', 'M': 'K', 'K': 'M', 'R': 'Y', 'Y': 'R', 'W': 'W', 'S': 'S',
'V': 'B', 'B': 'V', 'H': 'D', 'D': 'H', 'N': 'N', '-': '-' }
    #turns our string (s) into a list under variable 'letters'
    letters = list(s)
    #Iterate through the list 'letters' and for each 'base' capture the complement.
    #over-write the existing letter with the complement in list 'letters'
    letters = [basecomplement[base] for base in letters]
    #join the list elements to return a string.
    return ''.join(letters)
def revcom(s):
    '''Takes a string, and returns a reverse compliment of that string (as a string).
    Calls the function "complement".'''
    return complement(s[::-1])

some_string = 'AACCGGTT'
some_list = [1,2,3,4,5]

print(reverse(some_list))
print(complement(some_string))
print(revcom(some_string))
```

```python
## Q6 Parse fasta and produce a reverse compliment

fasta_file = '/home/andrew/Downloads/workshop_data.fasta'

## logic steps:
#1) reuse function from Q4
#2) Open the file
#3) iterate through the lines
#4) determine if correct line type
#5) if sequence, generate revcom
#6) print something...

with open(fasta_file, 'r') as f_in:
    for line in f_in:
        if not line.startswith('>'):
            print(revcom(line.strip()))
        else:
            # print the header with an adjusted title.
            print(f'{line.strip()}_revcom')
```

```python
## Q7 function to correct numerical order

numbers_list =['98', '67', '112', '4', '45', '78', '99', '100', '-1', '7', '0']

#logic step
#1) fix items in list so they are not strings

def correct_order(item):
    '''Takes a list, sorts the output and returns a sorted list.'''
    some_list = []
    for items in item:
        some_list.append(int(items))
    return sorted(some_list)

print(correct_order(numbers_list))
```

```python
## Q8 pull a random number from list

import random

print(random.choice(correct_order(numbers_list)))
```
